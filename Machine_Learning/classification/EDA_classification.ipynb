# Explanatory Data Analysis

### Dependencies

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read the Excel file
df_male = pd.read_excel('../datasets/male_data.xlsx')
df_female = pd.read_excel('../datasets/female_data.xlsx')
```

```python
# Calculate the number of male and female patients
num_male = len(df_male)
num_female = len(df_female)

# Data for the pie chart
labels = ['Male', 'Female']
sizes = [num_male, num_female]
colors = ['skyblue', 'lightcoral']
explode = (0.1, 0)  # explode the 1st slice (i.e. 'Male')

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Percentage of Male and Female Patients')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
def plot_osa_pie_chart(df):
    """
    Plot a pie chart based on the 'OSA' column in the dataset.
    """
    # Count the occurrences of each OSA category
    osa_counts = df['OSA'].value_counts()
    
    # Define labels and colors for the chart
    labels = ['No OSA', 'OSA']
    colors = ['#66c2a5', '#fc8d62']
    
    # Plot the pie chart
    plt.figure(figsize=(6, 6))
    plt.pie(osa_counts, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors, wedgeprops={'edgecolor': 'black'})
    plt.title('OSA Distribution')
    plt.show()
```

```python
plot_osa_pie_chart(df_male)
```

```python
plot_osa_pie_chart(df_female)
```
